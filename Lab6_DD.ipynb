```python
from efficient_apriori import apriori
import pandas as pd
from mlxtend.frequent_patterns import fpgrowth
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import fpmax
```

Zad 1

```python
db = [('Bread', 'Milk'),
                ('Bread', 'Diapers', 'Beer', 'Eggs'),
                ('Milk', 'Diapers', 'Beer', 'Cola'),
              ('Bread', 'Milk', 'Diapers', 'Beer'),
              ('Bread', 'Milk', 'Diapers', 'Cola'),
              ('Bread', 'Beer', 'Cola'),
              ('Bread', 'Milk', 'Eggs'),
              ('Bread', 'Milk'),
              ('Bread', 'Eggs'),
              ('Diapers', 'Beer', 'Cola'),
              ('Milk', 'Diapers', 'Beer', 'Cola'),
              ('Bread', 'Milk', 'Diapers', 'Beer', 'Eggs', 'Cola')]
```

```python
encoder = TransactionEncoder()
tmp = encoder.fit(db).transform(db)
df = pd.DataFrame(tmp, columns=encoder.columns_)
df
```

Zad 2

```python
items, rules = apriori(db, min_support=0.3)
rules
```

```python
fpgrowth(df, min_support=0.3, use_colnames=True)
```

Zad 4

```python
fpmax(df, min_support=0.3, use_colnames=True)
```
